#Gauss-Jordan Method

The following code features one primary method called "gaussJordan" that does the entirety of the Gauss-Jordan method. The Gauss-Jordan method takes place in three stages, and we will show an example below.

Starting equations:

$
x + y + 2z = -1  
$

$
x + 3y - 6z = 7  
$

$
2x - y + 2z = 0  
$

Starting matrix derived from equations:

<div style="text-align: left">
$
\begin{matrix}
\left[ \begin{array}{cccc}
1 & 1 & 2 & | & -1 \\
1 & 3 & -6 & | & 7 \\
2 & -1 & 2 & | & 0 \\
\end{array}
\right]
\end{matrix}
$
</div>

**Stage 1: transforming the matrix to create an upper triangular method**  

We found a valuable equation to transform the matrix to create a triangular matrix. We then implemented this equation into a method called "gJEquation." This method has two parameters and is used in both stage 1 and stage 3, and it is as follows. 

arr[j][k] - (arr[i][k] * (arr[j][i] / arr[i][i]) )

To implement the above equation, the "gJEquation" method must be called while in a double nested loop with specific conditions, which we will get into later. The following steps are gJEquation in action.

**Stage 1, Step 1**

row<sub>1</sub> - ( row<sub>0</sub> * ( (row<sub>1</sub>, column<sub>0</sub>) / (row<sub>0</sub>, column<sub>0</sub>) ) )

row<sub>1</sub> - ( row<sub>0</sub> * (1 / 1) )

[1,  3, -6,  7] - ( [1,  1,  2, -1] * (1 / 1) )

[1,  3, -6,  7] - [1,  1,  2, -1]

[0, 2, -8, 8]

**Stage 1, Step n**

row<sub>n</sub> - ( row<sub>1</sub> * ( (row<sub>n</sub>, column<sub>1</sub>) / (row<sub>1</sub>, column<sub>1</sub>) ) )

row<sub>n</sub> - ( row<sub>1</sub> * (2 / 1) )

[2, -1,  2,  0] - ( [1,  1,  2, -1] * (2 / 1) )

[2, -1,  2,  0] - [2, 2, 4, -2]

[0, -3,  -2,  2]

**Result:**

<div style="text-align: left">
$
\begin{matrix}
\left[ \begin{array}{cccc}
1 & 1 & 2 & | & -1 \\
0 & 2 & -8 & | & 8 \\
0 & -3 & -2 & | & 2 \\
\end{array}
\right]
\end{matrix}
$
</div>  

**Stage 1, Step 3**

row<sub>3</sub> - ( row<sub>2</sub> * ( (row<sub>3</sub>, column<sub>2</sub>) / (row<sub>2</sub>, column<sub>2</sub>) ) )

row<sub>3</sub> - ( row<sub>2</sub> * (3 / 2) )

[0, -3,  -2,  2] - ( [0, 2, -8, 8] * (3 / 2) )

[0, -3,  -2,  2] - [0, -3,  -12,  12]

[0.0, 0.0, -14.0, 14.0]

**Stage 1 Result**

\end{array}
\right]

The "gaussJordan" method has the ability to go further than three steps and is adaptable to the matrix size. This is why there is an option for a 5x6 two-dimensional array. We will elaborate further at the end.

**Stage 2: transforming the matrix for the Diagonal Matrix location values to be 1**

In this stage, we can use the following equation to transform the matrix in the diagonal position into "1":

row<sub>n</sub> = ( row<sub>n</sub> / (row<sub>n</sub>, column<sub>n</sub>) )

For example:

(Note: row<sub>1</sub> is the second array, and row<sub>0</sub> is the first array)

row<sub>1</sub> = ( row<sub>1</sub> / (row<sub>1</sub>, column<sub>1</sub>) )

row<sub>1</sub> = ( [0, 2, -8, 8] / (1 / 2) )

row<sub>1</sub> = [0, 1, -4, 4]

<div style="text-align: left">
$
\begin{matrix}
\left[ \begin{array}{cccc}
1 & 1 & 2 & | & -1 \\
0 & 1 & -4 & | & 4 \\
0 & 0 & 1 & | & -1 \\
\end{array}
\right]
\end{matrix}
$
</div>

**Stage 3: Transform matrix into a Diagonal Matrix**

This stage is very similar to stage 1 in that we transform row by row and column by column to get to the diagonal matrix. As we mentioned previously, it uses the same equation held within a double-nested loop to achieve the following results:

arr[j][k] - (arr[i][k] * (arr[j][i] / arr[i][i]) )

\right]
\end{matrix}
$

#Gauss-Jordan Code Example

````
# Array definitions
array1 = [ #array with simple answers (default)
         [1,  1,  2, -1],
         [1,  3, -6,  7],
         [2, -1,  2,  0]]

array2 = [# 2x2 case from below
         [1,  -2,  2],
         [2,  -4,  4]]

array3 = [# Test with many zeros - answers are not simple
         [1,  0,  1, -1],
         [0,  3, -0,  7],
         [2, 0,  0,  0]]

array4 = [# 4x5 test
         [1,  1,  2, -1, 2],
         [1,  3, -6,  7, 3],
         [1,  2, -4,  6, 3],
         [2, -1,  2,  0, 4]]

array5 = [# 5x6 test
         [1,  1,  2, -1, 2, 2],
         [1,  3, -6,  7, 3, 2],
         [1,  2, -4,  6, 3, 6],
         [4,  3,  2, -6, 1, 9],
         [2, -1,  2,  0, 4, 4]]

#---------------------------------------------------------------------

# Methods

def gJEquation(a, b, c, d):
  return a - ( b * (c / d) )

def gaussJordan(arr):

  showResults = True #@param {type:"boolean"}

  #Stage 1 Creating an upper triangular matrix
  val1 = 1

  for i in range (len(arr)):
    i_num = arr[i][i]

    for j in range (val1, len(arr)):
      j_num = arr[j][i]

      for k in range (len(arr[0])):
        arr[j][k] = gJEquation(arr[j][k], arr[i][k], j_num, i_num)

    val1 += 1

  if (showResults):
    for row in arr:
      print(row)
    print()

  #Stage 2 Transform rows for the Diagonal Matrix location values to be 1
  for i in range(len(arr)):
    i_num = arr[i][i]

    for j in range(len(arr[0])):
      arr[i][j] /= i_num
      #print(arr[i][j], i_num)

  if (showResults):
    for row in arr:
      print(row)
    print()

  #Stage 3 Transform rows to create a Diagonal Matrix
  val1 = len(arr)

  for i in range (1, len(arr)):
    i_num = arr[i][i]

    for j in range (i):
      j_num = arr[j][i]

      for k in range (len(arr[0])):
        arr[j][k] = gJEquation(arr[j][k], arr[i][k], j_num, i_num)

  val1 -= 1

  if (showResults):
    for row in arr:
      print(row)
    print()

  return arr

def sysOfEqAnswers(arr):
  tempArray = []

  gaussJordan(arr)

  for i in range(len(arr)):
    tempArray.append(arr[i][-1])

  return tempArray

#---------------------------------------------------------------------

print("Diagonal Matrix via Gauss-Jordan Method:\n ", gaussJordan( array1 ))
print()
print("System of equation answers via Gauss-Jordan Method:\n  ", sysOfEqAnswers( array1 ))
print()
print("Diagonal Matrix via Gauss-Jordan Method:\n ", gaussJordan( array5 ))
print()
print("System of equation answers via Gauss-Jordan Method:\n  ", sysOfEqAnswers( array5 ))
````

The most tedious part of this code was the conditional statements within the nested loops. The goal was to pull the correct elements and insert them into the equation:

Stage 1 boiled down to "turn the first column of every row, after the first row, into a zero." then "turn the second column of every row, after the second row, into a zero." and so on until a triangle is made. This became a game of getting the correct i, j, and k values from each loop.

Notice that these three groups are the same number of elements needed to be zeros in a 3x4 matrix. 


i is associated with the columns, j is associated with the rows, and k is associated with each element in a row.

*   The first column of the matrix, i is equal to zero.
*   The second column of the matrix, i is equal to one.

*   The first row of the matrix, j is equal to zero (array 0 is the first row of the matrix)
*   The second row of the matrix, j is equal to one

An important thing to remember is that in stage 1, the amount of elements that need to become zeros declines linearly.

#Limitations

A limitation that we had with this code was the following equations: 

$x − 2y = 2$  
$2x − 4y = 4$


The problem is that we run into a divide-by-zero error. Our method multiplies the first equation by, in this case, two. We then subtract the first equation from the second. The problem with this is that if we multiply the first equation by two, we are left with the same equation as the second. This results in the following, thus giving a divide by zero.

$2x − 4y = 4$  
$2x − 4y = 4$  

2 4 2  
0 0 0  

A way to get around this is to create an algorithm that can detect anomalies like this and then use a different method of solving for x and y. We can look at each element of the resulting matrix and compare the columns. If the result can create a divide-by-zero error, we use a different problem-solving method as a safeguard. Otherwise, we can continue with the algorithm.
